```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')  # 코랩 폰트
# plt.rc('font', family='Malgun Gothic')     # 윈도우 폰트
plt.rcParams['axes.unicode_minus'] = False
```

## 1.데이타 수집 및 전처리

```python
filename = '2017_202403온라인쇼핑몰_취급상품범위별_상품군별거래액_20240501125205.csv'
df = pd.read_csv(filename, encoding='utf-8')
df
```

```python
df.info()
```

```python
 df.describe()
```

```python
#필요없는 column삭제
df.drop(columns=['상품군별(2)'], axis=1, inplace=True)
df.isnull().sum()
```

```python
#columns변경
df.columns = ['상품군별','범위별','202401월','202402월','202403월','2017년','2018년','2019년','2020년','2021년','2022년','2023년']
df.head(10)
```

```python
# 범위별 '계'만 추출하기 위한 정렬
df=df.sort_values(by = '범위별' )
df.head(30)
```

```python
df['범위별'].value_counts()
```

```python
# 범위별 '계'만 추출
df.drop(df.index[24:], inplace=True)  #가전전자통신기기 total값만 적용
df
df.head(30)
```

```python
# 데이터 타입을 int로 변경
cols = ['2017년', '2018년', '2019년']
df[cols] = df[cols].astype(int)
df.info()
```

```python
#필요없는 column삭제
df.drop(columns=['범위별'], axis=1, inplace=True)
df.isnull().sum()
```

```python
df
```

```python
#2024년 합계 열 삽입
df.loc[:,'2024년3개월'] =df.loc[:,'202401월':'202403월'].sum(axis=1)
df
```

###년도별 총구매금액(2017-2024.03)

```python
#년도별 총구매금액
df_yeartotal=df.iloc[0:1,[0,4,5,6,7,8,9,10,11]]
df_yeartotal
df_yeartotal.index = df_yeartotal['상품군별']
del df_yeartotal['상품군별']
# df_yeartotal=df[df.상품군별=='합계']
df_yeartotal
df_yeartotal_t = df_yeartotal.T
df_yeartotalist=df_yeartotal_t
df_yeartotalist
```

```python
#년도별 총구매금액 막대바 그래프
x = df_yeartotalist.index
y = df_yeartotalist.합계
plt.bar(x, y,color='lightgreen')

plt.title('년도별 온라인쇼핑 총구매액추이', fontsize=15)     # 그래프 제목
plt.xlabel('년도별')     # x축 이름
plt.ylabel('구매액')     # y축 이름
plt.xticks(rotation=45)     # X축 레이블 회전
plt.show()
```

```python
# 년도별 열 정리
df_year=df.iloc[1:,[0,4,5,6,7,8,9,10,11]]
df_year
```

```python
#합계 열 삽입
df_year.loc[:,'총합계'] = df_year.loc[:,'2017년':'2024년3개월'].sum(axis=1)
df_yearlist=df_year.sort_values(by = '총합계' )
df_yearlist
```

```python
#온라인쇼핑 구매추이 파이그래프
# plt.pie(df_y.최근4년합계, labels=df_y.상품군별, autopct='%1.1f%%')
# plt.show()
# plt.pie(df_y['최근4년합계'], labels=df_y['상품군별'], autopct=lambda x: f'{x:1.1f}%')
# plt.show()

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
labels = df_yearlist['상품군별'].tolist()  # 상품군별 값들을 리스트로 변환
colors = sns.color_palette('hls', len(labels))
# my_colors = ['#2a9d8f', '#e9c46a', '#f4a261', '#e76f51']
plt.figure(figsize=(10, 5))  # 그래프 크기 설정
plt.pie(df_yearlist['총합계'], labels=labels, autopct=lambda x: f'{x:1.1f}%', colors=colors)
plt.title('온라인쇼핑 구매 추이(2017-2024.03)', fontsize=14)
# plt.axis('equal')  # 원형 모양 유지
plt.legend(labels, loc='best',bbox_to_anchor=(1.5,0.5,0.3,0.7) )  # 범례 위치 조정
plt.show()
```
